# 9. 조건에 따른 데이터 선택
-> 조건에 해당하는 데이터를 가지고 오는 작업

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col= '지원번호')
df
```

```python
df['키'] >= 184 # 학생들의 키가 184 이상인지 여부를 T/F를 통해 나타내줌
```

```python
filt = (df['키'] >= 184)
df[filt] # df에서 filt 조건을 가지고 데이터를 조회해 오는 것임.// 키가 184 이상인 사람만 나타내줌
```

```python
df[~filt] # filt 조건을 역으로 적용함. 즉 키가 184 이상이 아닌 사람을 나타내줌
```

```python
df[df['키'] >=184] # 조건을 변수에 저장하지 않고 바로 적은 것
```

```python
df.loc[df['키'] >=184, '수학'] # 키가 184 이상인 학생들의 수학점수만 가지고 오고 싶다.
#df.loc[row, col]
```

```python
df.loc[df['키'] >=184,['이름','수학','과학']] # 키가 184 이상인 학생들의 이름, 수학, 과학 점수를 가지고 오고 싶다.
```

## 다양한 조건
### & 조건 (and)

```python
# 북산고에서 키가 184 이상인 학생들의 이름과 국어 성적을 뽑고 싶다. (&조건 사용)
df.loc[ (df['키'] >= 184) & (df['학교'] == '북산고'), ['이름','국어']] 
```

### | 조건(or)

```python
df.loc[ (df['키'] <170) | (df['키'] >200), ['이름','국어']]  # 키가 170보다 작거나 또는 200보다 큰 사람의 이름과 국어 데이터를 불러옴
```

###  str 함수

```python
filt = df['이름'].str.startswith('송') # '송'이라는 글자로 시작하는 데이터를 뽑아옴
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 '태'가 들어있는 데이터를 뽑아옴
df[filt]
```

```python
df[~filt] # 이름에 '태'가 들어가지 않는 데이터를 추출
```

```python
langs = ['Python','Java']
filt = df['SW특기'].isin(langs) # sw특기가 파이썬이거나 자바인 사람을 뽑는것
df[filt]
```

```python
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt] 
# 위에서는 황태산이 나오지 않았음. 지금은 str.lower를 통해 대문자 PYTHON을 소문자로 바꿔서 뽑음
```

```python
filt = df["SW특기"].str.contains('Java',na = False) # na = True nan데이터에 대해서 트루로 판별하라 False로 하고 싶으면 False를 적으면 된다.
df[filt] # nan 데이터를 false로 처리함.// Java라는 글자를 포함하는 데이터를 가지고 옴.
```
