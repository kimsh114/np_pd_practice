```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col= '지원번호')
df
```

## 데이터 채우기 fillna

```python
df.fillna('') # 데이터 프레임 전체의 nan 데이터를 빈 칸으로 채운다. 
```

```python
import numpy as np
df['학교'] = np.nan # 학교 데이터를 넘파이를 통해 nan으로 채움
df
```

```python
df.fillna('모름') # 데이터 프레임 전체 nan에 대해 '모름'이라는 단어가 들어감 
```

```python
df.fillna('모름', inplace = True) # inplace를 통해 실제 데이터에 반영시킴
```

## 데이터 제외하기

### dropna

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col= '지원번호')
df
```

```python
df.dropna() # 전체 데이터 중 nan을 포함하는 데이터를 삭제 해버림
```

- axis : index or columns   //인덱스 또는 컬럼을 기준으로 삭제한다.
- how : any or all   // any: 데이터 중에 하나라도 nan이 있으면 데이터를 지움, all: 컬럼 또는 로우 전체가 nan일때 삭제

```python
df.dropna(axis = 'index', how = 'any') 
# nan이 하나라도 있는 row를 삭제함. 
# how의 디폴트는 any이다.
```

```python
df.dropna(axis = 'columns') # nan값이 하나라도 있는 컬럼을 삭제
```

```python
df['학교'] = np.nan # how = all을 하기 위해 학교 데이터를 넘파이를 이용해 전부 nan으로 변환
df
```

```python
df.dropna(axis = 'columns', how = 'all') # 컬럼 데이터 전체가 nan인 경우에만 삭제 how = all // 학교 컬럼 삭제됨
```
